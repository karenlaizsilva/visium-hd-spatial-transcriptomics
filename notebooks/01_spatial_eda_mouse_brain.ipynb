```python
import scanpy as sc
import pandas as pd
import matplotlib.pyplot as plt
import os

figures_path = '../figures'

# Settings for plots
sc.settings.figdir = figures_path
sc.set_figure_params(dpi=150, facecolor='white')
```

```python
# data loading
# path to the bin folder (8um or 16um)
data_dir = "../dados_10x/output_cloud/square_008um"

# complete Visium HD object (matrix + image)
adata = sc.read_visium(data_dir)
adata.var_names_make_unique()

print(f"Dataset loaded: {adata.n_obs} spots and {adata.n_vars} genes.")
```

```python
# Calculate standard quality metrics
sc.pp.calculate_qc_metrics(adata, inplace=True)

# Visualizing QC metrics with Violin Plots
sc.pl.violin(adata, ['n_genes_by_counts', 'total_counts'], 
             jitter=0.4, multi_panel=True, save="_qc_metrics.png")
```
